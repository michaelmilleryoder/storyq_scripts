# Make indicator columns for tags

```python
import pandas as pd
fpath = '/data/storyq/movie_plot_synopses.csv'
data = pd.read_csv(fpath)
data
```

```python
dummies = pd.get_dummies(data.tags.str.split(', ').apply(pd.Series).stack()).sum(level=0)
dummies
```

```python
modified = pd.concat([data, dummies], axis=1)
modified
```

```python
# Save out
modified.to_csv('/data/storyq/movie_plots.csv', index=False)
```
