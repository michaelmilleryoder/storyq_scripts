# Assemble tech product reviews

```python
import os
import json

dirpath = '/data/storyq/amazon-reviews-ml/json/train'
fpath = os.path.join(dirpath, 'dataset_en_train.json')
with open(fpath) as f:
    data = [json.loads(line) for line in f.read().splitlines()]

len(data)
```

```python
data[0].keys()
```

```python
# List products
from collections import Counter

cats = [entry['product_category'] for entry in data]
Counter(cats)
```

```python
tech_cats = ['wireless', 'pc', 'electronics']
tech_reviews = [entry for entry in data if entry['product_category'] in tech_cats]
len(tech_reviews)
```

```python
import pandas as pd

reviews = pd.DataFrame(tech_reviews)
reviews
```

```python
pd.set_option('display.max_colwidth', None)
reviews.loc[reviews.product_category == 'wireless', ['stars', 'review_title', 'review_body']].sample(100)
```

```python
reviews['stars'] = reviews['stars'].astype(int)
reviews
```

```python
# Filter to positive and negative, sample 1000
import numpy as np

selected = reviews.loc[(reviews['stars'] >= 4) | (reviews['stars'] <= 2),]
conditions = [(selected['stars']>=4), (selected['stars']<=2)]
choices = ['positive', 'negative']
selected['sentiment'] = np.select(conditions, choices)
selected
```

```python
selected['sentiment'].value_counts()
```

```python
pos = selected.loc[selected['sentiment']=='positive', ['review_body', 'sentiment']].sample(500)
neg = selected.loc[selected['sentiment']=='negative', ['review_body', 'sentiment']].sample(500)
out = pd.concat([pos, neg]).sample(frac=1)
out
```

```python
out.to_csv('/data/storyq/amazon_tech_review.csv', index=False)
```
